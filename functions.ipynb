1. What is the difference between a function and a method in Python?

    In Python, functions and methods are both used to perform a block of code when called, but they have a key difference based on how they are used and where they are defined.
    - A function is a block of code that is defined using the def keyword and is not associated with any object (unless it is inside a class, in which case it may be a method).
    - Example:
    def greet(name):
    return f"Hello, {name}!"
    - A method is a function that is associated with an object (i.e., it is defined inside a class). It automatically takes the instance (self) as the first parameter when it's called on an object.
    -Example:
    class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"
    g = Greeter()
    g.greet("Alice")  # 'greet' is a method here

7. What are the advantages of using generators over regular functions?

    Generators offer several practical advantages over regular functions, especially when you're working with large data, streams, or performance-critical tasks.Generators don’t store the entire sequence in memory.Regular functions with lists return all values at once, which can eat up memory.Generators can model things like infinite number series or event streams without crashing your system.A regular function can't do this — it would try to return an infinite list!.
    - Example:
    # List: stores all numbers in memory
    def make_list(n):
    return [i for i in range(n)]

    # Generator: yields numbers one by one
    def make_generator(n):
    for i in range(n):
        yield i

9.  Explain the purpose and usage of the `map()` function in Python?

    The map() function is used to apply a function to each item in an iterable (like a list, tuple, etc.), and returns a map object (an iterator) with the results.
    - Example:
    nums = [1, 2, 3, 4]

    # Square each number
    squared = map(lambda x: x ** 2, nums)

    print(list(squared))  # [1, 4, 9, 16]
10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

    map(), filter(), and reduce() are all powerful functional programming tools in Python — they help you process and transform data without writing explicit loops.
    - example of map reduce and filter:
    map:
    nums = [1, 2, 3, 4]
    squared = map(lambda x: x ** 2, nums)
    print(list(squared))  # [1, 4, 9, 16]
    reduce:
    from functools import reduce

    nums = [1, 2, 3, 4]
    total = reduce(lambda x, y: x + y, nums)

    filter:
    nums = [1, 2, 3, 4]
    evens = filter(lambda x: x % 2 == 0, nums)
    print(list(evens))  # [2, 4]
